```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
image = cv2.imread("ball.jpg")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.subplot(1,2,2)
plt.imshow(image_gray,'gray')
```

```python
hist_bins = 256
hist_range = (0,256)
plt.figure(2)
plt.hist(image_gray.ravel(), bins=hist_bins),plt.gca().set_xlim(hist_range)
```

```python
image_thres = image_gray < 100
plt.figure(2)
plt.imshow(image_thres,'gray')
```

```python
image_edge = cv2.Canny(image_gray,100,255)
plt.figure(3)
plt.imshow(image_edge,'gray')
```
